```python
import pandas as pd
```

```python
movie = pd.read_csv('https://gist.githubusercontent.com/tiangechen/b68782efa49a16edaf07dc2cdaa855ea/raw/0c794a9717f18b094eabab2cd6a6b9a226903577/movies.csv')
movie.head()
```

```python
movie[movie["Film"].str.contains("the")]["Film"]#trovare una parola in una Stringa
```

```python
movie[movie["Film"].str.endswith("on")] #visualizzare i film che finiscono con on
```

```python
movie[movie["Film"].str.startswith("T")] #visualizzare i film che iniziano con T
```

```python
movie.sort_values(by=["Film"])#visualizzare in ordine alfabetico i Film
```

```python
movie.sort_values(by=["Film"],ascending=False)#visualizzare in ordine decrescente i Film
```

ascending e pre impostato su true quindi per farlo al contrario si imposta su False

ordinare prima sul genere e poi sul nome del film

```python
movie.sort_values(by=["Genre","Film"],ascending=True)
```
